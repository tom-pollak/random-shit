```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('boosted-odds.csv')
df.dropna(how='any', inplace=True)
x = df['odds']
X = df['odds'].values.reshape(-1, 1)
y = df['boosted_odds']


reg = LinearRegression(fit_intercept=False).fit(X, y)
print("Regression score:", reg.score(X, y))
print(f"y = {reg.coef_[0]}x {reg.intercept_}")

y_new = np.round(df['odds'] * 1.2 - 1, 2)
plt.plot(X, y)
plt.plot(X, X * reg.coef_ + reg.intercept_, '--')
print(reg.coef_)
plt.show()
for i, j in zip(X, y):
    print(i, j)
```
